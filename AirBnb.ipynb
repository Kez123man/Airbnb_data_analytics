```python
import pandas as pd 

df = pd.read_csv('Airbnb_Data.csv')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df.columns
```

```python
df_1 = df.drop(columns=['host_has_profile_pic', 'first_review', 'last_review', 'thumbnail_url', 'zipcode'], axis=1)
```

```python
df_1.shape
```

```python
df_1.info()
```

```python
df_2 = df_1.dropna()
```

```python
df_2.shape
```

```python
df_2.describe()
```

```python
df_3 = df_2.drop_duplicates()
df_3.shape
```

```python
df_3['id'] = df_3['id'].astype(object)

df_3.dtypes
```

```python
plt.title('Histogram_Price_Graph')
sns.histplot(data=df_3['log_price'], bins=50)
plt.ylabel('Frequency')
```

```python
df_4 = df_3[df_3['review_scores_rating'] > 50]
plt.title('Histogram_Review_Score_Graph')
sns.histplot(data=df_4['review_scores_rating'], bins=50)
plt.ylabel('Frequency')
```

```python
df_4.groupby(by='property_type')['log_price'].median()
```

```python
city_med = df_4.groupby(by='city')['log_price'].median()
city_med.sort_values(ascending=False)
```

```python
df_4.groupby(by='neighbourhood')['log_price'].median()
```

```python
sns.barplot(data=df_4, x='city', y='log_price')
```

```python
sns.scatterplot(data=df_4, x='review_scores_rating', y='number_of_reviews', hue='city')
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_4, x='review_scores_rating', y='number_of_reviews', hue='bedrooms', palette='bright')
```

```python
sns.pairplot(data=df_4, vars= ['log_price', 'number_of_reviews', 'bedrooms', 'beds'], hue='city')
```

```python
# heat map: correlation of one variable for others

heat = df_4[['log_price', 'number_of_reviews', 'bedrooms', 'beds', 'latitude', 'longitude', 'number_of_reviews', 'review_scores_rating']]

sns.heatmap(data=heat, annot=True)
```
